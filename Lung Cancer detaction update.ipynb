# Transform the data into machine learning understanding (0,1) using MinMaxscaler()

```python
import sklearn.preprocessing as pp
Min_max=pp.MinMaxScaler()

ordinal_encoder = pp.OrdinalEncoder()
a = pf.iloc[:, :-1]
b = pf.iloc[:, -1]
```

# Correlation()

```python
import seaborn as sns
import matplotlib.pyplot as plt
corrMatrix = a.corr()
plt.figure(figsize=(14, 10))  # Adjust the figure size as needed
sns.heatmap(corrMatrix, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler,OrdinalEncoder

# Define your column transformer
transform_x = make_column_transformer(
    (MinMaxScaler(), ['Age', 'Gender', 'Air Pollution', 'Alcohol use', 'Dust Allergy',
       'OccuPational Hazards', 'Genetic Risk', 'chronic Lung Disease',
       'Balanced Diet', 'Obesity', 'Smoking', 'Passive Smoker', 'Chest Pain',
       'Coughing of Blood', 'Fatigue', 'Weight Loss', 'Shortness of Breath',
       'Wheezing', 'Swallowing Difficulty', 'Clubbing of Finger Nails',
       'Frequent Cold', 'Dry Cough', 'Snoring'])
)
```

```python
transform_x
```

```python
x = transform_x.fit_transform(a)
ordinal_encoder = OrdinalEncoder()
y = ordinal_encoder.fit_transform(np.array(b).reshape(-1, 1))
```

```python
x
```

```python
y
```

# Applying Algorithms On Data For Training and Testing prediction

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.1)
```

# SVM()

```python
S = svm.SVC()
```

```python
S.fit(x_train,y_train)
```

```python
S_training_accuracy = S.score(x_train,y_train)
```

```python
S_training_accuracy
```

```python
y_pred = S.predict(x_test)
```

```python
S_testing_accuracy = accuracy_score(y_test,y_pred)
```

```python
S_testing_accuracy
```

```python
plt.bar(['Training'],[S_training_accuracy],label = 'Training')
plt.bar(['Testing'],[S_testing_accuracy],label = 'Testing')

plt.xlabel('Accuracy')
plt.ylabel('Score')
plt.title('SVM() Training & Testing Accuracy')
plt.legend()
plt.show()
```

# KNN()

```python
y_pred = knn.predict(x_test)
```
